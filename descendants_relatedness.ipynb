```python
def genome_similarity(person_a, person_b):
    matching = 0
    for gene in range(20_000):
        if person_a['genome'][gene] == person_b['genome'][gene]:
            matching += 1
    return matching / 20_000 * 100
```

```python
max_generation = max([x['generation'] for x in simulation.people.values()])
last_generation = [x['name'] for x in simulation.people.values() if x['generation'] == max_generation]
combinations = [*itertools.combinations(last_generation,2)]
#for combo in combinations:
relatedness = []
for combination in combinations:
    relatedness.append(genome_similarity(simulation.people[combination[0]],
                                         simulation.people[combination[1]]))
relatedness
```

```python
similarities = dict()
for generation in range(2,7):
    last_generation = [x['name'] for x in simulation.people.values() if x['generation'] == generation]
    combinations = [*itertools.combinations(last_generation,2)]
    #for combo in combinations:
    relatedness = []
    for combination in combinations:
        relatedness.append(genome_similarity(simulation.people[combination[0]],
                                             simulation.people[combination[1]]))
    similarities[f"generation {generation}"] = {'mean': np.mean(relatedness),
                                               'min': min(relatedness),
                                               'max': max(relatedness),
                                               'std': np.std(relatedness)}
    print(f"{generation} shared: mean: {round(np.mean(relatedness),2)}")
```
